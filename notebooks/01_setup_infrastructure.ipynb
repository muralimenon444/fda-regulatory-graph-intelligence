```python
# -------------------------------------------------------------------------------------------
# NOTEBOOK: 01_setup_infrastructure
# ROLE: Unity Catalog Infrastructure Setup (Catalog, Schemas, Volume)
# -------------------------------------------------------------------------------------------

# Deterministic configuration (per master context)
CATALOG_NAME = "hc_regulatory_sandbox"
SCHEMAS = ["bronze_ingestion", "silver_entities", "gold_analytics"]  # Fixed: gold_analytics (not gold_graph)
VOLUME_NAME = "landing_zone"

print(f"Initializing infrastructure: {CATALOG_NAME}\n")

# Create Catalog
spark.sql(f"CREATE CATALOG IF NOT EXISTS {CATALOG_NAME}")
print(f"  ✓ Catalog: {CATALOG_NAME}")

# Create Medallion Schemas
for schema in SCHEMAS:
    spark.sql(f"CREATE SCHEMA IF NOT EXISTS {CATALOG_NAME}.{schema}")
    print(f"  ✓ Schema: {CATALOG_NAME}.{schema}")

# Create Managed Volume (Landing Zone)
spark.sql(f"CREATE VOLUME IF NOT EXISTS {CATALOG_NAME}.bronze_ingestion.{VOLUME_NAME}")
print(f"  ✓ Volume: {CATALOG_NAME}.bronze_ingestion.{VOLUME_NAME}")

print(f"\n{'='*60}")
print(f"✅ Infrastructure setup complete")
print(f"{'='*60}")
```
