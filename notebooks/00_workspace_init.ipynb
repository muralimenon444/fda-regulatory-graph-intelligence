```python
# -------------------------------------------------------------------------------------------
# NOTEBOOK: 00_workspace_init
# ROLE: Workspace Initialization - Creates deterministic folder hierarchy
# -------------------------------------------------------------------------------------------

import os

# Deterministic repository root (per master context)
REPO_ROOT = "/Workspace/Repos/muralimenon444@gmail.com/fda-regulatory-graph-intelligence"

# Lean professional directory hierarchy
FOLDERS = [
    "notebooks",        # Sequential execution notebooks
    "app",              # Streamlit frontend
]

print(f"Initializing workspace at: {REPO_ROOT}\n")

for folder in FOLDERS:
    path = os.path.join(REPO_ROOT, folder)
    try:
        os.makedirs(path, exist_ok=True)
        print(f"  ✓ {folder}")
    except Exception as e:
        print(f"  ✗ {folder}: {str(e)}")
        raise

print(f"\n{'='*60}")
print(f"✅ Lean workspace structure initialized")
print(f"{'='*60}")
```
